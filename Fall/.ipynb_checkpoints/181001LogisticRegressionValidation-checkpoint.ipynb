```python
# Using Logistic Regression to predict if a person would have survived the titanic, given their fare and age
# Data group 1: survivors
# Data group 2: dead
# Basically, i need to find the average of logits given a parameter, and test the dataset 
# randomly select 1/10 of the data and test it aside (or give each data point a value 1-10 for the category group 
# such that 10% of the total data falls in each category [0 will be value for unassigned to a group]) 
# sklearn.fit to find the coefficients and intercept for everything except the first set of data
# then calculate the probabilities of each of the testing datapoints using model.predict.proba (but do probability being survivor)
# next, find the -log(probabilty) if dead and -log(1-probability) if survivor
# sum and average these to find the average cost and save in a list
# repeat for the remaining groups
# average everything in the list 
```

```python
# Import necessary libraries and read csv
from sklearn import linear_model # for fitting our model
import sklearn
import pandas as pd
import matplotlib.pyplot
import numpy

# force numpy not to use scientific notation, to make it easier to read the numbers the program prints out
numpy.set_printoptions(suppress=True)

titanic_data_full = pd.read_csv('/Users/nikpate/Documents/Programming/MachineLearningClass20182019/Fall/titanic_dataset.csv')
```

```python
# Set up data and set parameters here
titanic_data = titanic_data_full[['fare','age','sex','pclass','sibsp','survived']].dropna()
titanic_data['random_group'] =0
sections = 10
```

```python
def assign_samples(data,sections):
    #Take data and randomly assign them groups 1-10
    
    points_per_section = int(data.shape[0]/sections)
    sampled_data = pd.DataFrame({'fare':[],'age':[],'sex':[],'pclass':[],'sibsp':[],'survived':[],'random_group':[]})
    sampled_data['survived'] = sampled_data['survived'].astype(int)
    for i in range(10):

        current_sample = data.sample(n=points_per_section)
        data = data.drop(current_sample.index)
        current_sample['random_group'] = i+1
        sampled_data = pd.concat([sampled_data,current_sample])
        
    return sampled_data.sort_index().reset_index(drop=True)
```

```python
def build_model(training_data):
    #uses training data to train model
    
    training_classes = list(training_data['survived'])
    training_inputs = list(zip(training_data['fare'],training_data['age'],training_data['sex'],training_data['pclass'],training_data['sibsp']))
    model = linear_model.LogisticRegression()
    model.fit(training_inputs, training_classes)
    
    return model
```

```python
def test_proba(testing_data,model):
    #test probabilities
    
    testing_inputs = list(zip(testing_data['fare'],testing_data['age'],testing_data['sex'],testing_data['pclass'],testing_data['sibsp']))
    testing_data['prediction'] = list(list(zip(*model.predict_proba(testing_inputs)))[1])
    
    return testing_data
```

```python
def sample_proba(data,section):
    #given our data (sampled) and specifying a section, 
    # build a model with training data and then find the probabilities for the test data
    
    training_data = data[data['random_group'] != section]
    testing_data = data[data['random_group'] == section]
    
    model = build_model(training_data)
    
    return test_proba(testing_data,model)
```

```python
def roc_value(testing_data,descision_boundary):
    #find the True positive rate and false positive rate
    
    true_positive = testing_data[(testing_data['survived'] == 1) & (testing_data['prediction'] >= descision_boundary)].shape[0]
    false_negative = testing_data[(testing_data['survived'] == 1) & (testing_data['prediction'] < descision_boundary)].shape[0]
    false_positive = testing_data[(testing_data['survived'] == 0) & (testing_data['prediction'] >= descision_boundary)].shape[0]
    true_negative = testing_data[(testing_data['survived'] == 0) & (testing_data['prediction'] < descision_boundary)].shape[0]
    
    true_positive_rate = true_positive/(false_negative+true_positive)
    false_positive_rate = false_positive/(false_positive+true_negative)
    
    return [false_positive_rate,true_positive_rate]
    
```

```python
def make_roc_data(testing_data):
    #find tpr and fpr for all percents 1-100, graph them, and calculate auc
    resolution = 500
    x = []
    y = []
    area_under_curve = []
    
    for i in range(0,resolution):
        coordinates = roc_value(testing_data,i/resolution)
        x.append(coordinates[0])
        y.append(coordinates[1])

    for i in range(1,resolution):
        y1=y[i]
        x1=x[i-1]
        x2=x[i]
        area_under_curve.append((x1-x2)*y1)
    
    auc_val = sum(area_under_curve)
    
    return (x,y,auc_val)
```

```python
def plot_roc_curve(roc_data):
    
    fig = matplotlib.pyplot.figure()
    fig.suptitle('ROC Curve')
    matplotlib.pyplot.xlabel('FPR')
    matplotlib.pyplot.ylabel('TPR')
    matplotlib.pyplot.plot(roc_data[0], roc_data[1])
    
    X = numpy.linspace(0, 1)
    Y = [i for i in list(zip(X))]
    matplotlib.pyplot.plot(X, Y,c="black")
    
    matplotlib.pyplot.show()
    
```

```python
def roc_curve(testing_data):
    #find coordinates, plot curve, and plot 50-50 line
    
    roc_data = make_roc_data(testing_data)
    print('auc: ' + str(roc_data[2]))
    
    # uncomment line below to show graphs
    plot_roc_curve(roc_data)
    
    return roc_data[2]
```

```python
def prc_value(testing_data,descision_boundary):
    #find the True positive rate and false positive rate
    
    true_positive = testing_data[(testing_data['survived'] == 1) & (testing_data['prediction'] >= descision_boundary)].shape[0]
    false_negative = testing_data[(testing_data['survived'] == 1) & (testing_data['prediction'] < descision_boundary)].shape[0]
    false_positive = testing_data[(testing_data['survived'] == 0) & (testing_data['prediction'] >= descision_boundary)].shape[0]
    true_negative = testing_data[(testing_data['survived'] == 0) & (testing_data['prediction'] < descision_boundary)].shape[0]
    
    if (false_positive+true_positive != 0):
        precision = true_positive/(false_positive+true_positive)
    else:
        precision = None
    recall = true_positive/(false_negative+true_positive)
    
    return [recall,precision]
    
```

```python
def make_prc_data(testing_data):
    #find tpr and fpr for all percents 1-100, graph them, and calculate auc
    resolution = 500
    x = []
    y = []
    area_under_curve = []
    
    for i in range(0,resolution):
        coordinates = prc_value(testing_data,i/resolution)
        x.append(coordinates[0])
        y.append(coordinates[1])

    for i in range(1,resolution):
        y1=y[i]
        x1=x[i-1]
        x2=x[i]
        if y1 != None:
            area_under_curve.append((x1-x2)*y1)
    
    auc_val = sum(area_under_curve)
    
    return (x,y,auc_val)
```

```python
def plot_prc_curve(prc_data):
    
    fig = matplotlib.pyplot.figure()
    fig.suptitle('PRC Curve')
    matplotlib.pyplot.xlabel('Recall')
    matplotlib.pyplot.ylabel('Precision')
    matplotlib.pyplot.plot(prc_data[0], prc_data[1])
    
    matplotlib.pyplot.show()
    
```

```python
def prc_curve(testing_data):
    #find coordinates, plot curve, and plot 50-50 line
    
    prc_data = make_prc_data(testing_data)
    print('auc: ' + str(prc_data[2]))
    
    # uncomment line below to show graphs
    plot_prc_curve(prc_data)
    
    return prc_data[2]
```

```python
def run_x_fold(sample_data,sections):
    #find area under curve for each section and average
    auc = []
    testing_data = sample_proba(sample_data,1)
    
    for i in range(1,10):
        
        testing_data = sample_proba(sample_data,i)
        #auc.append(roc_curve(testing_data))
        auc.append(prc_curve(testing_data))
    
    error = (sum(auc)/len(auc))
    
    print('Mean error: ' + str(error))
    
    return error
```

```python
sample_data = assign_samples(titanic_data,sections)



run_x_fold(sample_data,sections)
```
