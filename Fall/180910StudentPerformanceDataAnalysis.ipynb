```python
#What is the correlation/influence of failures on G1?
student_data['failures'].corr(student_data['G1'])
```

```python
#What is the correlation/influence of family size on G1?
print(student_data['G1'][(student_data['famsize'] == 'LE3')].mean())
print(student_data['G1'][(student_data['famsize'] == 'GT3')].mean())
```

```python
#What is the correlation/influence of absences on G1?
student_data['absences'].corr(student_data['G1'])
```

```python
#What variable or set of variables can you find that most accurately predict G3?
student_data.corr()['G3'].abs().sort_values(ascending=False)
#While past grades are the best predictors of G3, failures, Mother's education, age, and Father's education also serve as important indicators.
```
